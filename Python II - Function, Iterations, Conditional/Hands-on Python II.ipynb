```python
def print_harga_dibayar(jumlah_barang, harga_barang, ppn=1.1):
  harga_dibayar = jumlah_barang * harga_barang
  harga_dibayar = harga_dibayar * ppn  # ppn
  print(int(harga_dibayar))

print_harga_dibayar(5, 5000, 1.11)
print_harga_dibayar(2, 2000, 1.11)
print_harga_dibayar(10, 50000, 1.11)
# print(harga_dibayar)
```

```python
def pangkat_dua(angka):
  return angka * angka

p2 = pangkat_dua(2)
p3 = pangkat_dua(3)
p4 = pangkat_dua(4)
print(p2, p3, p4)
```

```python
def pertambahan_after_pangkat_dua(list_angka):
  hasil = 0
  for item in list_angka:
    pangkat_dua = a(item)
    print(hasil, '+', pangkat_dua)
    hasil += pangkat_dua
  return hasil

print(pertambahan_after_pangkat_dua([1, 2, 3]))
```
